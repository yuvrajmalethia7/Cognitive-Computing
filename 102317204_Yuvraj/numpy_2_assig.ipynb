```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

add = gfg.sum()
print("sum of all elements:", add)

row = gfg.sum(axis=1)
print("row wise add:", row)

col = gfg.sum(axis=0)
print("column wise add:", col)
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print("Sorted array:", sorted_array)

indices_sorted = np.argsort(array)
print("Indices of sorted array:", indices_sorted)

smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)

largest_5 = np.sort(array)[-5:]
print("5 largest elements:", largest_5)
```

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == array.astype(int)]
print(integer_elements)

float_elements = array[array != array.astype(int)]
print(float_elements)
```

```python
initials_ascii_sum = ord('A') + ord('B') 
X = initials_ascii_sum

sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Sales dataset:", sales)

tax = ((X % 5) + 5) / 100
print("Tax rate:", tax)

sales_tax = sales * (1 + tax)
print("Sales after applying tax:", sales_tax)

discount = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print("Sales after discount:", discount)

weekly = np.tile(sales, (3, 1))
weekly = weekly * (1 + 0.02 * np.arange(3)).reshape(3, 1)
print("Weekly sales with increase:", weekly)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x ** 2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Y = log(|x| + 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.tight_layout()
plt.show()
```
